```python
total_frota = df_frota['Frota'].sum()
```

```python
df_frota['Porcentagem'] = (df_frota['Frota']/total_frota) * 100

df_frota
```

```python
fonte_consulta_uf = {
    'detran_mg': 'MG', 'detran_rs': 'RS','der_ba': 'BA', 'detran_rn': 'RN','detran_sc': 'SC', 'git_parana': 'PR',
    'detran_rj': 'RJ', 'detran_mt': 'MT','detran_ma': 'MA', 'detran_am': 'AM','detran_es': 'ES', 'detran_rr': 'RR',
    'detran_ba': 'BA'
}
```

```python
df_consultas.loc[df_consultas['+todos'] == 1, 'Expectativa'] = total_frota


df_consultas
```

```python
def calculo_expectativa_consultas(fonte_consulta):
    if fonte_consulta in fonte_consulta_uf:
        return df_frota.loc[df_frota['UF'] == fonte_consulta_uf[fonte_consulta], 'Frota'].iloc[0]
    else:
        return 0
```

```python
df_consultas.loc[df_consultas['Expectativa'].isna(), 'Expectativa'] = df_consultas.loc[df_consultas['Expectativa'].isna(), 'orgaos_homologados'].apply(calculo_expectativa_consultas)


df_consultas
```

```python
df_consultas['Porcentagem'] = (df_consultas['consulta_semanal']/df_consultas['Expectativa']) *100


df_consultas
```
